## -- Model.pkl file viewer --


```python
import pickle
import joblib

# Path to your model.pkl
path = "model.pkl"

# Try loading the model using pickle first, then joblib as fallback
try:
    with open(path, "rb") as f:
        model1 = pickle.load(f)
    print("✅ Model loaded successfully using pickle.")
except Exception as e:
    print("⚠️ Pickle failed, trying joblib...")
    try:
        model1 = joblib.load(path)
        print("✅ Model loaded successfully using joblib.")
    except Exception as e2:
        print("❌ Unable to load model file.")
        print("Error:", e2)
        model1 = None

# Show model details if loaded
if model1 is not None:
    print("\nModel type:", type(model1))
    
    # If it's a Keras or TensorFlow model
    try:
        model1.summary()
    except:
        # For sklearn or other models
        print("\nModel object preview:\n", model1)
```

```python
import pandas as pd
import pickle

# Load the pickle file
path = "model.pkl"

with open(path, "rb") as file:
    data = pickle.load(file)

# If it's a dictionary (like your case), convert to DataFrame
if isinstance(data, dict):
    df = pd.DataFrame([data])
else:
    # If it’s something else (like a list or object), convert accordingly
    df = pd.DataFrame([str(data)], columns=["Model_Object"])

# Save to CSV
csv_path = "model_contents.csv"
df.to_csv(csv_path, index=False)

print(f"✅ Model content saved to {csv_path}")
print(df)
```
